```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
T = 1000
dt = 0.1 
total_steps = int(T/dt)
V_th=12
I = 14
t_ref = 5
C_m = 10
t_rest = 0
```

```python
def update_membrane_potential(V,I):
    V = V +  (I / C_m ) * dt
    return V
```

```python
voltage_values = []
time_values = []

for _time in range(total_steps):
    if _time > t_rest:
        V = update_membrane_potential(V, I)  # update membrane potential 
    else:
        V = 0
        
    if V >= V_th:
        V = V_th + 2
        t_rest = _time + t_ref 
    
        
    voltage_values.append(V)
    time_values.append(_time)
    
plt.plot(time_values[0:1000], voltage_values[0:1000], color="b")
plt.title("IF Neuron")
plt.xlabel("time")
plt.ylabel("Membrane Potential")
plt.show()
```
